```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
```

```python
#reading the file into a dataframe
df = pd.read_csv('Project 2 Data.csv')

#visual assessment of dataset
df.head()
```

```python
#programmatic assessment of dataset
df.info()
```

```python
#converting variables to appropriate datatype
df[['Private_Attorney', 'Marital_Status', 'Gender', 'Severity']] = df[['Private_Attorney', 'Marital_Status', 'Gender', 'Severity']].astype('category')
```

```python
#checking the summary statistics of the numerical variables
df.describe()
```

```python
#plotting the distribution of age
df.hist('Age', bins=30)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Claimant Age');
```

```python
#plotting the distribution of severity
#count the frequencies of each category
severity_counts = df['Severity'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
severity_counts.plot(kind='bar', color='teal')
plt.xlabel('Severity Level')
plt.ylabel('Count')
plt.title('Distribution of Severity Levels')
plt.xticks(rotation=0)  #rotates X-Axis Ticks by 45-degrees
plt.show();
```

```python
#plotting the distribution of claim payment amount
sns.histplot(df['Amount'], bins=30)
plt.title('Distribution of Claim Payment Amount')
plt.xlabel('Amount ($)')
plt.ylabel('Frequency');
```

```python
#plotting the distribution of representation by a private attorney
palette = sns.color_palette()

ax = sns.countplot(x='Private_Attorney', data=df)
plt.title('Distribution of Private Attorney Representation')
plt.xlabel('Private Attorney')
plt.ylabel('Frequency')

legend_labels = ['No', 'Yes']
handles = [plt.Rectangle((0,0),1,1, color=palette[i]) for i in range(len(legend_labels))]
ax.legend(handles, legend_labels, title="Private Attorney")

plt.show()
```

```python
#performing linear regression between 'Age' and 'Amount'
regression_result = linregress(df['Age'], df['Amount'])

#extracting the slope and p-value from the regression result
slope = regression_result.slope
p_value = regression_result.pvalue

#checking if the relationship is statistically significant
is_significant = p_value < 0.05

slope, p_value, is_significant
```

```python
#plotting the regression line
slope = -534.9110454177785
intercept = regression_result.intercept

#scatter plot of Age vs Amount
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Amount', data=df)

#calculating the regression line using the slope and intercept
y_values = slope * df['Age'] + intercept

#adding the regression line to the plot
sns.lineplot(df['Age'], y_values, color='red')

#setting plot title and labels
plt.title('Age of Claimant vs Claim Payment Amount with Regression Line')
plt.xlabel('Age of Claimant')
plt.ylabel('Claim Payment Amount ($)')

plt.show()
```

```python
#performing linear regression between 'Severity' and 'Amount'
regression_result_severity = linregress(df['Severity'], df['Amount'])

#extracting the slope and p-value from the regression result
slope_severity = regression_result_severity.slope
p_value_severity = regression_result_severity.pvalue

#checking if the relationship is statistically significant
is_significant_severity = p_value_severity < 0.05

slope_severity, p_value_severity, is_significant_severity
```

```python
#converting severity back to a numerical column
df['Severity'] = df['Severity'].astype(int)

#scatter plot of the data points
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Severity', y='Amount', data=df)

#calculating the regression line
x_values = df['Severity']
y_values = regression_result_severity.slope * x_values + regression_result_severity.intercept

#adding the regression line to the plot
sns.lineplot(x=x_values, y=y_values, color='red')

#setting the plot title and labels
plt.title('Severity of Damage vs Claim Payment Amount with Regression Line')
plt.xlabel('Severity of Damage')
plt.ylabel('Claim Payment Amount ($)')

plt.show()
```

```python
#splitting the data into two groups based on Private_Attorney representation
grouped_data = df.groupby('Private_Attorney')

#performing linear regression for each group
regression_results = {}

for name, group in grouped_data:
    result = linregress(group['Severity'], group['Amount'])
    regression_results[name] = result

#preparing the plot
plt.figure(figsize=(10, 8))

#plotting each group with its regression line
for name, group in grouped_data:
    #scatter plot for each group
    sns.scatterplot(x='Severity', y='Amount', data=group, label=f'Private Attorney = {name}')

    #regression line for each group
    x_values = group['Severity']
    y_values = regression_results[name].slope * x_values + regression_results[name].intercept
    sns.lineplot(x=x_values, y=y_values, color=('green' if name == 1 else 'blue'))

#setting the plot title and labels
plt.title('Relationship between Severity and Payment Amount by Private Attorney Representation')
plt.xlabel('Severity')
plt.ylabel('Claim Payment Amount ($)')

plt.legend()
plt.show()

#outputting the slope and p-value for each group
regression_results_attorney = {
    "With Private Attorney": regression_results[1],
    "Without Private Attorney": regression_results[0]
}
regression_results_attorney
```

```python
#create boxplots of the payment amount distributed by severity levels for each group
plt.figure(figsize=(14, 7))

#boxplot for claimants with a private attorney
plt.subplot(1, 2, 1)
sns.boxplot(x='Severity', y='Amount', data=df[df['Private_Attorney'] == 1])
plt.title('Payment Amount by Severity (With Private Attorney)')
plt.xlabel('Severity')
plt.ylabel('Amount')

#boxplot for claimants without a private attorney
plt.subplot(1, 2, 2)
sns.boxplot(x='Severity', y='Amount', data=df[df['Private_Attorney'] == 0])
plt.title('Payment Amount by Severity (Without Private Attorney)')
plt.xlabel('Severity')
plt.ylabel('Amount')

plt.tight_layout()
plt.show()
```
